```python
drop_list = ['year']
df_main['Age'] = 2020 - df_main['year']
```

```python
df_main['selling_price(lacs)'] = df_main['selling_price'] / 100000
drop_list.append('selling_price')
```

```python
df_main['km_driven(lacs)'] = df_main['km_driven'] / 100000
drop_list.append('km_driven')
```

```python
# mileage as float
mil_kmph = []
for i in range(df_main['mileage'].shape[0]):
    mil_kmph.append(float(df_main['mileage'].iloc[i].split()[0]))
df_main['mileage(kmph)'] = mil_kmph
drop_list.append('mileage')
```

```python
# engine as float
engine_1000cc = []
for i in range(df_main['engine'].shape[0]):
    engine_1000cc.append(int(df_main['engine'].iloc[i].split()[0]) / 1000)
df_main['engine(1000cc)'] = engine_1000cc
drop_list.append('engine')
```

```python
# max_power as float
max_power_bhp = []
for i in range(df_main['max_power'].shape[0]):
    max_power_bhp.append(float(df_main['max_power'].iloc[i].split()[0]))
df_main['max_power(bhp)'] = max_power_bhp
drop_list.append('max_power')
```

```python
drop_list.extend(['torque'])
```

```python
df_main.drop(drop_list, axis= 1, inplace= True)
```

```python
Y = df_main['selling_price(lacs)']
X = df_main[['km_driven(lacs)', 'engine(1000cc)', 'seats', 'max_power(bhp)', 'Age', 'mileage(kmph)']]
```

```python
featSel = GenericUnivariateSelect(f_classif, mode='k_best', param=3)
X_new = featSel.fit_transform(X, Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, Y, test_size=0.2, random_state=1)
print("x train: \t", X_train.shape)
print("x test: \t", X_test.shape)
print("y train: \t", y_train.shape)
print("y test: \t", y_test.shape)
```

```python
def car_model_prediction(model, model_name):
    model.fit(X_train, y_train)
    
    # R2 score of train set
    y_pred_train = model.predict(X_train)
    R2_train_model = round(r2_score(y_train,y_pred_train), 2)
    
    # R2 score of test set
    y_pred_test = model.predict(X_test)
    R2_test_model = round(r2_score(y_test,y_pred_test), 2)
   
    
    cross_val = cross_val_score(model ,X_train ,y_train ,cv= 5)
    cv_mean = round(cross_val.mean(), 2)
    
    print("Train R2-score :", R2_train_model)
    print("Test R2-score :", R2_test_model)
    print("Train CV scores :", cross_val)
    print("Train CV mean :", cv_mean)
    
    # Plotting Graphs 
    # Residual Plot of train data
    fig, ax = plt.subplots(1,2, figsize= (10,4))
    ax[0].set_title('Residual Plot of Train samples')
    sns.distplot((y_train - y_pred_train), hist= False,ax= ax[0])
    ax[0].set_xlabel('y_train - y_pred_train')
    
    # Y_test vs Y_train scatter plot
    ax[1].set_title('y_test vs y_pred_test')
    ax[1].scatter(x= y_test, y= y_pred_test)
    ax[1].set_xlabel('y_test')
    ax[1].set_ylabel('y_pred_test')
    
    plt.show()
```

```python
lr = LinearRegression()
car_model_prediction(lr, "Linear_regressor.pkl")
```
