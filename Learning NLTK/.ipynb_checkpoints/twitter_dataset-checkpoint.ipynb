```python
train_data, test_data = train_test_split(data_used, test_size=0.2, random_state=1)
```

```python
train_pos = train_data[ train_data['sentiment'] == 'Positive']
# train_pos = train_pos['text']
train_neg = train_data[ train_data['sentiment'] == 'Negative']
# train_neg = train_neg['text']
train_neutral = train_data[train_data['sentiment'] == 'Neutral']
# train_neutral = train_neutral['text']
```

```python
def wordcloud_draw(data, color = 'black'):
    words = ' '.join(data)
    cleaned_word = " ".join([word for word in words.split()
                            if 'http' not in word
                                and not word.startswith('@')
                                and not word.startswith('#')
                                and word != 'RT'
                            ])
    wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color=color,
                      width=2500,
                      height=2000
                     ).generate(cleaned_word)
    plt.figure(1,figsize=(13, 13))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
    
print("Positive words")
wordcloud_draw(train_pos['text'],'white')
```

```python
print("Negative words")
wordcloud_draw(train_neg['text'])
```

```python
print('Neutral words')
wordcloud_draw(train_neutral['text'])
```

```python
# clean words
stopwords_set = set(nltk.corpus.stopwords.words('english'))
tweets = []
punct_remove = str.maketrans('', '' ,string.punctuation)
for index, row in train_data.iterrows():
    # regularise cases and decrease, and elimante strings of length 0, 1, 2
    words_filtered = [e.lower() for e in row.text.split() if len(e) >= 3]
    words_cleaned = [word.translate(punct_remove) for word in words_filtered 
                     if 'http' not in word 
                     and not word.startswith('@')
                     and not word.startswith('#')
                     and word != 'RT']
    # remove stop words, remove strings less than length 3 that emerge due to using word.translate
    words_without_stopwords = [word for word in words_cleaned 
                                if not word in stopwords_set and len(word) >= 3]
    tweets.append((words_without_stopwords, row.sentiment))
# print(tweets)
```

```python
test_pos = test_data[ test_data['sentiment'] == 'Positive']
# test_pos = test_pos['text']
test_neg = test_data[ test_data['sentiment'] == 'Negative']
# test_neg = test_neg['text']
```

```python
# get words present in the tweets
word_list = [y for x in tweets for y in x[0]]
```

```python
# count number of times a word is there in word_list
word_count = nltk.FreqDist(word_list)
unq_words = word_count.keys()
```

```python
wordcloud_draw(unq_words)
```
